## Data Preprocessing


```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

cat_features = ['job','marital','education','targeted','default','housing','loan','contact','month']

X = pd.get_dummies( bank_df, columns=cat_features, drop_first=True)
```

```python
sc = MinMaxScaler()
a = sc.fit_transform(bank_df[['salary']])
b = sc.fit_transform(bank_df[['balance']])
c = sc.fit_transform(bank_df[['day']])
d=sc.fit_transform(bank_df[['duration']])
e=sc.fit_transform(bank_df[['campaign']])
f = sc.fit_transform(bank_df[['pdays']])
g = sc.fit_transform(bank_df[['previous']])
h = sc.fit_transform(bank_df[['pdays_no_contact']])
```

```python
X['salary'] = a
X['balance'] = b
X['day']=c
X['duration']=d
X['campaign']=e
X['pdays'] = f
X['previous'] = g
X['pdays_no_contact']=h
```

```python
X.shape
```

## Resampling

```python
sns.countplot('response', data=bank_df).set_title('Class Distribution Before Resampling')
```

```python
X_no = X[X.response == 0]
X_yes = X[X.response == 1]
```

```python
X_yes_upsampled = X_yes.sample(n=len(X_no), replace=True, random_state=42)
print(len(X_yes_upsampled))
```

```python
X_upsampled = X_no.append(X_yes_upsampled).reset_index(drop=True)
print(len(X_upsampled))
```

```python
sns.countplot('response', data=X_upsampled).set_title('Class Distribution After Resampling')
```

# ML model

### train test and split

Now it's time to do a train test split, and train our model!


```python
from sklearn.model_selection import train_test_split
```

```python
x = X_upsampled.drop(['response'], axis=1) #features (independent variables)
y = X_upsampled['response'] #target (dependent variable)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.25,random_state=70)
```

## Logistic Regression 


```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

lr.fit(X_train,Y_train)
```

```python
lr_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 
```

```python
Y_test[:10]
```

```python
lr_pred[:10]
```

```python
confusion_matrix(Y_test,lr_pred)
```

```python
accuracy_score(Y_test,lr_pred)*100 
```

```python
print(classification_report(Y_test,lr_pred))
```

##### So accuracy of our model with logistic regression is approx 82%.

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)

rfc.fit(X_train, Y_train)

rf_pred = rfc.predict(X_test)
```

```python
Y_test[:10]
```

```python
rf_pred[:10]
```

```python
confusion_matrix(Y_test,rf_pred)
```

```python
accuracy_score(Y_test,rf_pred)*100 
```

```python
print(classification_report(Y_test,rf_pred))
```

##### So we get the accuracy of our model is 97% which is much better than Logistic Regression algorithm model.
